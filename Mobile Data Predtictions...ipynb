```python
columns_numeric= ['battery_power', 'blue', 'clock_speed', 'dual_sim', 'fc', 'four_g',
       'int_memory', 'm_dep', 'mobile_wt', 'n_cores', 'pc', 'px_height',
       'px_width', 'ram', 'sc_h', 'sc_w', 'talk_time', 'three_g',
       'touch_screen', 'wifi']
```

# Imputation

```python
from sklearn.impute import SimpleImputer
```

```python
# NUMERIC DATA
#Train data
imputer_num = SimpleImputer(strategy='median')
imputer_num.fit(df_train[columns_numeric])
df_train[columns_numeric] = imputer_num.transform(df_train[columns_numeric])

# Test data
df_test[columns_numeric] = imputer_num.transform(df_test[columns_numeric])
```

```python
df_train.isna().sum().sum()
```

# Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# Scaling
# Train data
scaler = MinMaxScaler()
scaler.fit(df_train[columns_numeric])
df_train[columns_numeric]= scaler.transform(df_train[columns_numeric])

# Test data 
df_test[columns_numeric] = scaler.transform(df_test[columns_numeric])
```

# Model Building

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_RF = RandomForestClassifier(n_estimators=100)
model_RF.fit(df_train[columns_numeric],df_train['price_range'])
```

```python
model_RF.score(df_train[columns_numeric],df_train['price_range'])
```

```python
# Make predictions
yp = model_RF.predict(df_test[columns_numeric])
```

```python
len(yp)
```

```python
yp=yp.astype(int)
```

```python
df_submission =pd.DataFrame({'Id':df_test['Id'],'price_range':yp})
```
